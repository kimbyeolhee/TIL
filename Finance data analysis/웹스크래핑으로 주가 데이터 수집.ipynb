## 함수 

```python
def get_day_list(item_code, page_no):
    """
    일자별 시세를 페이지 별로 수집
    """
    url = f"https://finance.naver.com/item/sise_day.nhn?code={item_code}&page={page_no}"
    
    header = {'user-agent':"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/90.0.4430.93 Safari/537.36"}

    response = requests.get(url, headers = headers)
    html = bs(response.text)
    table = html.select("table")
    
    temp = pd.read_html(str(table))
    temp = temp[0].dropna()
    
    return temp
```

```python
get_day_list(item_code, 10)
```

# 반복문을 통한 전체 일자 데이터 수집

```python
import time
# 시작 페이지
page_no = 1
# 데이터를 저장할 빈 변수
item_list=[]

while True:
    print(page_no)
    temp = get_day_list(item_code, page_no)
    
    item_list.append(temp)
    
    page_no = page_no + 1
    time.sleep(0.1) # 일방적인 데이터 요청은 네트워크에 부담이 되어` DDOS 공격으로 오해받을 가능성

    # if temp.shape[0] < 10 :  전체 페이지를 긁어오는 것
    if page_no > 10:
        break;
        
```

# 총 정리: 전체 함수로 만들기

```python
def get_item_list(item_code, item_name ):
    """
    일별 시세를 수집하는 함수
    """
    # 시작 페이지
    page_no = 1
    # 데이터를 저장할 빈 변수
    item_list=[]

    while True:
        temp = get_day_list(item_code, page_no)

        item_list.append(temp)

        page_no = page_no + 1
        time.sleep(0.1) # 일방적인 데이터 요청은 네트워크에 부담이 되어` DDOS 공격으로 오해받을 가능성

        # if temp.shape[0] < 10 :  전체 페이지를 긁어오는 것
        if page_no > 10:
            break;
    
    df=pd.concat(item_list)
    df["종목코드"] = item_code
    df["종목명"] = item_name
    cols = ['종목코드','종목명','날짜','종가','전일비','시가','고가','저가','거래량']
    df = df[cols]
    
    return df
```

```python
item_code = "326030"
item_name = "SK바이오팜"

get_item_list(item_code, item_name)
```
