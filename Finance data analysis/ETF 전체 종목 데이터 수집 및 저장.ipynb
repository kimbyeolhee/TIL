- JSON형태로 출력할려고 하면 오류 발생
- txtx로 보면 window.__jindo2_callback._7568 값이 앞에 있는 것을 볼 수 있다. JSON형태는 { 로 시작해야하는데..
- url을 보면 뒤에 같은 내용이 붙어있는 것을 확인, 지워서 다시 실행해보자

```python
url = "https://finance.naver.com/api/sise/etfItemList.nhn?etfType=0&targetColumn=market_sum&sortOrder=desc"
url
```

```python
response = requests.get(url)
response
```

## JSON 타입으로 데이터 받기

```python
response.json()
```

```python
etf_json = response.json()
```

## JSON에서 원하는 데이터 찾기

```python
etfItemList = etf_json["result"]["etfItemList"]
etfItemList
```

```python
len(etfItemList)
```

## 데이터프레임 변환

### JSON to pandas DataFrame

```python
df =pd.DataFrame(etfItemList)
df.shape
```

```python
df.head()
```

# 파일 저장

## 파일명 만들기

```python
import datetime

today = datetime.datetime.today()
today = today.strftime("%Y-%m-%d")

today
```

```python
# f-string 방식
file_name = f"etf_{today}_raw.csv"
file_name
```

## 파일 저장 및 불러오기

```python
df.to_csv(file_name, index=False)
```

```python
pd.read_csv(file_name, dtype={"itemcode":np.object}) # itemcode 앞의 0이 숫자로 읽혀서 생략되어 출력되므로 object로 출력되게 함
```
