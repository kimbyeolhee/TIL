# 데이터 전처리

## etfTabName 만들기

```python
# etf Tab Code를 구분하기 위한 list 생성

etfcode = """전체
국내 시장지수
국내 업종/테마
국내 파생
해외 주식
원자재
채권
기타"""

etfcode
```

```python
# split로 나누어 list형 데이터로 만듬
etf_tab_name = etfcode.split("\n")
```

```python
# 번호에 따른 Tab Name 반환 함수

def find_etf_tab_name(no):
    return etf_tab_name[no]

find_etf_tab_name(2)
```

```python
df["etfTabCode"].apply(find_etf_tab_name) #apply는 series, DataFrame에 사용가능, map은 series 형태에만 적용가능
```

```python
df["etfTabName"] = df["etfTabCode"].map(lambda x : etf_tab_name[x])
```

```python
# etfTabName 칼럼 확인

df.head()
```

```python
df.loc[df["etfTabCode"] == 1, ["itemname","etfTabName"]]
```

## 컬럼 명 변경

```python
cols = df.columns.tolist()
cols
```

```python
col_name = """종목코드
탭코드
종목명
현재가
등락구분
전일비
등락률
순자산가치(NAV)
3개월수익률
거래량
거래대금(백만)
시가총액(억)
유형"""

col_name =  col_name.split("\n")
col_name
```

```python
# 칼럼 영문명과 한글명을 딕셔너리 형태로 변환
dict(zip(cols, col_name))
```

```python
df.columns = col_name
df.head()
```

## 파생변수

### 브랜드명

```python
df["종목명"] # 맨 앞이 브랜드명
```

```python
df["종목명"].str.split(" ", expand=True) # expand=True 데이터프레임형태로 
```

```python
df["브랜드"] = df["종목명"].str.split(" ", expand=True)[0]
df[["종목명","브랜드"]].head()
```

```python
df["브랜드"].unique()
```

### 인버스

```python
# 인버스ETF는 지수가 하락하면 오히려 수익률이 오르도록 설계된 상품
df["인버스"] = df["종목명"].str.contains("인버스")
df["인버스"].value_counts()
```

### 레버리지

```python
# 수익이 날 떄 더 수익이 나고 손해일때는 더 손해가 남
df["레버리지"] = df["종목명"].str.contains("레버리지")
df["레버리지"].value_counts(normalize=True) #비율
```

### 헷지

```python
# 확정되지 않은 자산을 확정된 자산으로 편입하는 과정, 위험자산의 가격변동을 제거하는 것
df["헷지"] = df["종목명"].str.endswith("H)")
df[["종목명","헷지"]].head()
```
