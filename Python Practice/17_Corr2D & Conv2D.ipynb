# Corr2D
<br>
<font size="5">
    $$ \vec{x} \otimes \vec{f} [i,j] =
    \sum_{u=-k}^{k} \sum_{v=-k}^{k}
    x[i+u, j+v] f[u, v]$$
</font>


<br>
<font size="5">
    $$ L_{o} = L_{i} - L_{f} + 1 $$
</font>

<br>

k=1을 대입해서 비교해보면 180도 뒤집는다는 것이 어떤 의미인지 파악 가능

## 미리 구현

```python
import random

filter_size = 3
H, W = 5,6

data = list()
for _ in range(H):
    data.append([random.randint(0, 10) for _ in range(W)])

corr_filter = list()
for _ in range(filter_size):
    corr_filter.append([random.randint(0,10) for _ in range(filter_size)])

print("Input Signal: ")
for data_row in data:
    print(data_row)
print("\nCorrelational Filter: ")
for corr_row in corr_filter:
    print(corr_row)
print('===========================================================')

valid_H = H - filter_size + 1  # 3
valid_W = W - filter_size + 1  # 4

corr = list()

for h in range(valid_H):
    corr_vec = list()
    for w in range(valid_W):
        window = list()
        
        for inner_h in range(filter_size):
            window.append(data[h+inner_h][w:w+filter_size])
        
        corr_sum = 0
        for h in range(filter_size):
            for w in range(filter_size):
                corr_sum += window[h][w] * corr_filter[h][w]
        corr_vec.append(corr_sum)
    corr.append(corr_vec)

print(corr)
```

## 복습 구현

```python
import random

filter_size = 3
H, W = 5,6

data = list()
for _ in range(H):
    data.append([random.randint(0, 10) for _ in range(W)])

corr_filter = list()
for _ in range(filter_size):
    corr_filter.append([random.randint(0,10) for _ in range(filter_size)])

print("Input Signal: ")
for data_row in data:
    print(data_row)
print("\nCorrelational Filter: ")
for corr_row in corr_filter:
    print(corr_row)
print('===========================================================')

valid_H , valid_W = H - filter_size + 1 , W - filter_size + 1
corr_mat = list()

for h in range(valid_H):
    corr_vec = list()
    for w in range(valid_W):
        window = list()
        
        for inner_h in range(filter_size):
            window.append(data[h+inner_h][w:w+filter_size])
        
        corr = 0
        for window_row , filter_row in zip(window, corr_filter):
            for window_val, filter_val in zip(window_row, filter_row):
                corr += window_val*filter_val
        corr_vec.append(corr)
    corr_mat.append(corr_vec)

for vec in corr_mat:
    print(vec)
```

# Conv2D
<br>
<font size="5">
    $$ \vec{x} * \vec{f} [i,j] =
    \sum_{u=-k}^{k} \sum_{v=-k}^{k}
    x[i+u, j+v] f[-u, -v]$$
</font>


<br>
<font size="5">
    $$ L_{o} = L_{i} - L_{f} + 1 $$
</font>

```python
import random

filter_size = 3
H, W = 5,6

data = list()
for _ in range(H):
    data.append([random.randint(0, 10) for _ in range(W)])

conv_filter = list()
for _ in range(filter_size):
    conv_filter.append([random.randint(0,10) for _ in range(filter_size)])

print("Input Signal: ")
for data_row in data:
    print(data_row)
print("\nConvolutional Filter: ")
for conv_row in conv_filter:
    print(conv_row)
print('===========================================================')

valid_H , valid_W = H - filter_size + 1 , W - filter_size + 1
conv_mat = list()

for h in range(valid_H):
    conv_vec = list()
    for w in range(valid_W):
        window = list()
        for inner_h in range(valid_H):
            window.append(data[h+inner_h][w:w+filter_size])
            
        print(window, '---', conv_filter)
        
        window_flatten, filter_flatten = list(), list()
        for inner_h in range(filter_size):
            for inner_w in range(filter_size):
                window_flatten.append(window[inner_h][inner_w])
                filter_flatten.append(conv_filter[inner_h][inner_w])
        print(window_flatten, '---', filter_flatten)
        
        filter_flatten = filter_flatten[::-1]
        
        conv = 0
        for window_val, filter_val in zip(window_flatten, filter_flatten):
            conv += window_val * filter_val
        conv_vec.append(conv)
    conv_mat.append(conv_vec)

print('======')
for conv in conv_mat:
    print(conv)
```
