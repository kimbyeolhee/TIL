# 미리 구현 

```python
class_names = ['Korean', 'English', 'Math']
scores = [[10, 15, 30],
         [20, 45, 20],
         [30, 35, 40],  
         [40, 30, 50]]

# setting
n_student = len(scores)
n_class = len(scores[0]) 

meta_data = dict()
interested_data = ['sum', 'mean', 'M_idx', 'm_idx']

# format
for class_name in class_names:
    meta_data[class_name] = dict()
    for i_data in interested_data:
        meta_data[class_name][i_data] = 0

# calculate sum, get Max, min
for class_name in class_names:
    meta_data[class_name]['Max'] = None
    meta_data[class_name]['Min'] = None

for s_idx, s_scores in enumerate(scores):
    for c_idx, c_score in enumerate(s_scores):
        meta = meta_data[class_names[c_idx]]
        meta['sum'] += c_score
        
        if  meta['Max'] == None or meta['Max'] < c_score:
            meta['Max'] = c_score
            meta['M_idx'] = s_idx
        if meta['Min'] == None or meta['Min'] > c_score:
            meta['Min'] = c_score
            meta['m_idx'] = s_idx

# calculate mean
for c_idx in range(n_class):
    meta = meta_data[class_names[c_idx]]
    meta['mean'] = meta['sum'] / n_student

print("meta_data_dictionary")
for k,v in meta_data.items():
    print(f"subject : {k}", v)

# calculate mean_substraction , scre_normalization
ms_mat = list()
norm_mat = list()

for s_idx, s_score in enumerate(scores):
    tmp_ms = list()
    tmp_norm = list()
    for c_idx, c_score in enumerate(s_score):
        meta = meta_data[class_names[c_idx]]
        
        tmp_ms.append(c_score - meta['mean'])
        tmp_norm.append((c_score - meta['Min'])/(meta['Max'] - meta['Min']))
        
    ms_mat.append(tmp_ms)
    norm_mat.append(tmp_norm)

print("Mean Subtraction\t\tNormalization")
for ms_vec, norm_vec in zip(ms_mat, norm_mat):
    print(ms_vec,'\t',norm_vec)
```

# 복습 구현

```python
# problem setting
class_names = ['Korean', 'English', 'Math']
scores = [[10, 65, 90],
          [20, 45, 20],
          [30, 35, 40],
          [40, 30, 50]]

# define what we need for mean subtraction & normalization
interested_data = ['sum', 'mean', 'M_idx', 'm_idx']

# 초기화
n_student = len(scores)
n_class = len(scores[0])

meta_data = dict()

for class_name in class_names:
    meta_data[class_name] = dict()
    
    for i_data in interested_data:
        meta_data[class_name][i_data] = 0
        meta_data[class_name]['Max'] = None
        meta_data[class_name]['min'] = None

# mean subtraction & normalization을 구하기 위해 meta_data 값 계산
for s_idx, s_scores in enumerate(scores):
    for c_idx, c_score in enumerate(s_scores):
        meta = meta_data[class_names[c_idx]]
        meta['sum'] += c_score
        
        if meta['Max'] == None or meta['Max'] < c_score:
            meta['Max'] = c_score
            meta['M_idx'] = s_idx

        if meta['min'] == None or meta['min'] > c_score:
            meta['min'] = c_score
            meta['m_idx'] = s_idx
            
for class_idx, class_name in enumerate(class_names):
    meta = meta_data[class_name]
    meta['mean'] = meta['sum'] / n_student

for k, v in meta_data.items():
    print(f"subject: {k}=> ",v )

# get mean subtraction & normalization
ms_mat = list()
norm_mat = list()

for s_idx, s_scores in enumerate(scores):
    ms_vec = list()
    norm_vec = list()
    
    for c_idx, c_score in enumerate(s_scores):
        meta = meta_data[class_names[c_idx]]
        
        ms_vec.append(c_score - meta['mean'])
        norm_vec.append((c_score - meta['min']) / (meta['Max'] - meta['min']))
        
    ms_mat.append(ms_vec)
    norm_mat.append(norm_vec)

print("mean subtraction \t normalization")
for ms_vec, norm_vec in zip(ms_mat, norm_mat):
    print(ms_vec,'\t',norm_vec)
```
