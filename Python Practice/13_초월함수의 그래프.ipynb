# Sin, Cos Graphs

```python
import math
import matplotlib.pyplot as plt

PI = math.pi

# make linear space
low, high = 0, 4*PI
n_data = 100
interval = (high - low) / (n_data-1)

linspace = [low + interval*i for i in range(n_data)]

# list for sin, cos
sin = [math.sin(data) for data in linspace]
cos = [math.cos(data) for data in linspace]

# graphs
fig, ax = plt.subplots(figsize=(10,5))
ax.plot(linspace, sin, label = 'sin')
ax.plot(linspace, cos, label = 'cos')
ax.legend(fontsize = 10)
```

# Exp, Log Graphs

```python
import matplotlib.pyplot as plt
import math

# linear space
low, high = 0.01 , 1
n_data = 100
interval = (high - low) / (n_data - 1)

linspace = [low + interval*data_idx for data_idx in range(n_data)]

# list
exp = [math.exp(data) for data in linspace]
log = [math.log(data) for data in linspace]

# graph
fig, axes = plt.subplots(2, 1, figsize = (10,5))
axes[0].plot(linspace, exp)
axes[1].plot(linspace, log)
for ax_idx, ax in enumerate(axes.flat):
    ax.axvline(x=0, color='k', linestyle=':')
    ax.axhline(y=0, color='k', linestyle=':')
    ax.tick_params(labelsize = 15)
```

# Sigmoid Graph
math.exp를 이용하여 sigmoid 함수의 graph를 그려봅니다.<br>
<font size="5">
    $$\sigma(x) = \frac{1}{1 + e^{-x}}$$
</font>


```python
import matplotlib.pyplot as plt
import math

# linear space
low, high = -10, 10
n_data = 100
interval = (high - low) / (n_data - 1)

linspace = [low + interval*data_idx for data_idx in range(n_data)]

# list
sigmoid = [1 / (1 + math.exp(-data)) for data in linspace]

# graph
fig, ax = plt.subplots(figsize = (10,5))
ax.plot(linspace, sigmoid)
ax.grid()
```

# Tanh Graph
$$sinh(x) = \frac{e^{x} - e^{-x}}{2}$$
$$cosh(x) = \frac{e^{x} + e^{-x}}{2}$$
<font size="5">
    $$tanh(x) = 
    \frac{sinh(x)}{cosh(x)} = 
    \frac{e^{x} - e^{-x}}{e^{x} + e^{-x}}$$
</font>

```python
import matplotlib.pyplot as plt
import math

# linear space
low, high = -10, 10
n_data = 100
interval = (high - low) / (n_data - 1)

linspace = [low + interval*data_idx for data_idx in range(n_data)]

# list
tanh = list()
for data in linspace:
    numer = math.exp(data) - math.exp(-data)
    denom = math.exp(data) + math.exp(-data)
    tanh.append(numer / denom)

# graph
fig, ax = plt.subplots(figsize = (10,5))
ax.plot(linspace, tanh)
ax.grid()
```

# ReLU Graph
<font size="5">
    $$ReLU(x) = max(0, x)$$
</font>

```python
import matplotlib.pyplot as plt
import math

# linear space
low, high = -10, 10
n_data = 100
interval = (high - low) / (n_data - 1)

linspace = [low + interval*data_idx for data_idx in range(n_data)]

# list
relu = list()
for data in linspace:
    if data >= 0:
        relu.append(data)
    else:
        relu.append(0)

# graph
fig, ax = plt.subplots(figsize = (10,5))
ax.plot(linspace, relu)
ax.grid()
```

# Information Graph
정보이론에서 다루는 information graph <br>

<font size="5">
    $$I(p) = -log(p) = log\bigg( \frac{1}{p} \bigg), \quad where \; 0 < p < 1$$
</font>

```python
import math
import matplotlib.pyplot as plt

# set the linear space
low, high = 0.001, 1
n_data = 100

interval = (high - low)/(n_data - 1)
linspace = [low + data_idx*interval for data_idx in range(n_data)]

# list
information = [- math.log(p) for p in linspace]

# graph
fig, ax = plt.subplots(figsize = (10,5))
ax.plot(linspace, information)
ax.grid()
ax.set_xlabel("Probability", fontsize=20)
ax.set_ylabel("Information", fontsize=20)
ax.tick_params(labelsize=15)
```

# Binary Entropy Graph

<font size="5">
    $$H(x) = -\big( plog(p) + (1-p)log(1-p) \big)$$
</font>

```python
import math
import matplotlib.pyplot as plt

# set the linear space
low, high = 0.001, 0.999
n_data = 100

interval = (high - low)/(n_data - 1)
linspace = [low + data_idx*interval for data_idx in range(n_data)]

# list
binary_entropy = list()
for p in linspace:
    entropy = -(p*math.log(p) + (1-p)*math.log(1-p))
    binary_entropy.append(entropy)

# graph
fig, ax = plt.subplots(figsize = (10,5))
ax.plot(linspace, binary_entrophy)
ax.grid()
ax.set_xlabel("Probabiltiy", fontsize = 20)
ax.set_ylabel("Binary Entrophy", fontsize = 20)
ax.tick_params(labelsize = 15)
```
