# Making Unit Vectors

```python
import numpy as np
from termcolor import colored
np.set_printoptions(precision=2)

# generate test vectors
n_element = 4
u = np.random.uniform(0, 10, (n_element, ))
v = np.random.uniform(0, 10, (n_element, ))
print(f"u: {u}\nv: {v}\n")

# calculate and print the norms
u_len = np.sum(np.square(u))**0.5
v_len = np.sum(np.square(v))**0.5
print(colored("Before unitization",'blue'))
print(f"len of u: {u_len.round(2)}")
print(f"len of v: {v_len.round(2)}\n")

# unitize the vectors
u_unit = u/u_len
v_unit = v/v_len

# calculate and print the norms After unitization
u_len = np.sum(np.square(u_unit))**0.5
v_len = np.sum(np.square(v_unit))**0.5
print(colored("After unitization",'blue'))
print(f"len of u: {u_len.round(2)}")
print(f"len of v: {v_len.round(2)}")
```

# Maing Unit Vectors(Matrix)

```python
import numpy as np
from termcolor import colored
np.set_printoptions(precision=2)

# generate test vectors
n_vector, n_element = 3, 4
U = np.random.uniform(0, 10, (n_vector, n_element))
print(f"U: \n{U}\n")

# calculate and print norms Before unitization
U_len = np.sqrt(np.square(U).sum(axis=1)).reshape(-1, 1)
print(colored("Before unitization", 'blue'))
print(f"len of U: \n{U_len}\n")

# unitize the vectors
U_unit = U / U_len

# calculate and print norms After unitization
U_len = np.sqrt(np.square(U_unit).sum(axis=1)).reshape(-1,1)
print(colored("After unitization", 'blue'))
print(f"len of U: \n{U_len}\n")
```

# Euclidean Distance (Vector-wise)

```python
import numpy as np
np.set_printoptions(precision=2)

# generate test vectors
n_vector, n_element = 3, 4
U = np.random.uniform(0, 10, (n_vector, n_element))
V = np.random.uniform(0, 10, (n_vector, n_element))
print(f"U: {U.shape}\n{U}")
print(f"V: {V.shape}\n{V}\n") 

e_dists = np.sqrt(np.sum(np.square(U - V), axis=1))
print(f"Euclidean distances: {e_dists.shape}\n{e_dists}")
```

# Euclidean Distance (Fully-connected)

```python
import numpy as np
np.set_printoptions(precision=2)

# generate test vectors
n_vector1, n_vector2 = 3, 4
n_element = 4
U = np.random.uniform(0, 10, (n_vector1, n_element))
V = np.random.uniform(0, 10, (n_vector2, n_element))
print(f"U: {U.shape}\n{U}")
print(f"V: {V.shape}\n{V}\n") 

U_ = np.expand_dims(U, axis=0)
V_ = np.expand_dims(V, axis=1)
print(U_.shape, V_.shape)

e_dists = np.square(U_ - V_)
e_dists = np.sqrt(np.sum(e_dists, axis=-1)) # -1이여야 (1,3,**4**) 4 기준으로 더해줌

print(f"Eclidean distance: {e_dists.shape}")
```

```python
import numpy as np
from termcolor import colored
np.set_printoptions(precision=2)

# generate test vectors
n_element = 4
u = np.random.uniform(0, 10, (n_element, ))
v = np.random.uniform(0, 10, (n_element, ))
print(f"u: {u}\nv: {v}\n")

# unitize vectors
u_len = np.sum(np.square(u))**0.5
u_unit = u/u_len
v_len = np.sum(np.square(v))**0.5
v_unit = v/v_len

# dot product
cos = np.sum(u_unit * v_unit)  # dot product = cos value
print(f"dot product: {cos}")
```

# Cosine Similarlity (Vector-wise)

```python
import numpy as np
np.set_printoptions(precision=2)

# generate test vectors
n_vector, n_element = 3, 4
U = np.random.uniform(0, 10, (n_vector, n_element))
V = np.random.uniform(0, 10, (n_vector, n_element))
print(f"U: {U.shape}\n{U}")
print(f"V: {V.shape}\n{V}\n") 

U_norms = np.sqrt(np.sum(np.square(U), axis=1)).reshape(-1, 1)
U_unit = U / U_norms
V_norms = np.sqrt(np.sum(np.square(V), axis=1)).reshape(-1, 1)
V_unit = V / V_norms

# calculate cos similarities
similarity = np.sum(U_unit * V_unit, axis=1)
print(f"cosine similarities: \n{similarity}")
```

# Cosine Similarlity (Fully-connected)

```python
import numpy as np
np.set_printoptions(precision=2)

# generate test vectors
n_vector1, n_vector2 = 3, 4
n_element = 4
U = np.random.uniform(0, 10, (n_vector1, n_element))
V = np.random.uniform(0, 10, (n_vector2, n_element))
print(f"U: {U.shape}\n{U}")
print(f"V: {V.shape}\n{V}\n") 

U_norms = np.sqrt(np.sum(np.square(U), axis=1)).reshape(-1, 1)
U_unit = U / U_norms
V_norms = np.sqrt(np.sum(np.square(V), axis=1)).reshape(-1, 1)
V_unit = V / V_norms

U_ = np.expand_dims(U_unit, axis=0)
V_ = np.expand_dims(V_unit, axis=1)
print(U_.shape, V_.shape)

sims = np.sum(U_ * V_ , axis=-1)
print(f"cosine similarities: \n{sims}")
```
