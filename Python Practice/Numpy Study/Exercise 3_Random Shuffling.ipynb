# Shuffling a Vector

```python
import numpy as np

a = np.arange(10)
print(a)

np.random.shuffle(a)
print(a)
```

# Shuffling a Matrix

```python
import numpy as np

M = np.arange(15).reshape(5, -1)
print(M)

np.random.shuffle(M)
print(M)
```

# Shuffling indices

```python
import numpy as np
from termcolor import colored

N = 10
a = np.linspace(0, 10, N)
b = np.linspace(0, 100, N)
print(colored("before shuffling", 'blue'))
print(f"a: \n{a}")
print(f"b: \n{b}\n")

indicies = np.arange(N)
np.random.shuffle(indicies)

a_shuffle = a[indicies]
b_shuffle = b[indicies]

print(colored("After shuffling", 'blue'))
print(f"a_shuffle: \n{a_shuffle}")
print(f"b_shuffle: \n{b_shuffle}\n")
```

# Shuffling X, Y Data

```python
import numpy as np
from termcolor import colored

N, n_feature = 6, 5
n_class = 100

x_data = np.random.normal(0, 1, (N, n_feature))
y_data = np.random.randint(0, n_class, N)

print(colored("before Shuffling", 'blue'))
print(f"x_data: {x_data.shape}\n{x_data}")
print(f"y_data: {y_data.shape}\n{y_data}\n")

indices = np.arange(N)
np.random.shuffle(indices)

x_data = x_data[indices]
y_data = y_data[indices]

print(colored("After Shuffling", 'blue'))
print(f"x_data: {x_data.shape}\n{x_data}")
print(f"y_data: {y_data.shape}\n{y_data}")
```

# Shuffling and Getting Minibatches

랜덤하게 32개씩 뽑는 것이라 데이터가 중복될 가능성이 있음

```python
import numpy as np

N, n_feature = 200, 5
n_class =100
batch_size = 32
iters = 8

x_data = np.random.normal(0, 1, (N, n_feature))
y_data = np.random.randint(0, n_class, N)

for iter in range(iters):
  indices= np.arange(N)
  np.random.shuffle(indices)
  batch_indices = indices[:batch_size] # random한 32개 인덱스 추출

  x_batch = x_data[batch_indices]
  y_batch = y_data[batch_indices]

  print("iter: ", iter)
  print(f"X/Y: {x_batch.shape} / {y_batch.shape}")
```

데이터 셋을 전체 다 확인하는 경우의 코드

```python
import numpy as np

N, n_feature = 200, 5
n_class =100
batch_size = 32
iters = 8

x_data = np.random.normal(0, 1, (N, n_feature))
y_data = np.random.randint(0, n_class, N)

indices= np.arange(N)
np.random.shuffle(indices)

for iter in range(iters):
  batch_indices = indices[batch_size * iter : batch_size * (iter+1)] # random한 32개 인덱스 추출

  x_batch = x_data[batch_indices]
  y_batch = y_data[batch_indices]

  print("iter: ", iter)
  print(f"X/Y: {x_batch.shape} / {y_batch.shape}")
```
