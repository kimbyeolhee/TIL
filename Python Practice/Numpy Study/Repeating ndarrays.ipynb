# **Element-wisely Repetition**

## np.repeat

원소 별로 반복

```python
import numpy as np

x = 3

rep = np.repeat(x, 2)
print("x: ", x)
print("np.repeat(x, 2): ", rep)
```

```python
import numpy as np

x = np.array([1, 2, 3])

rep = np.repeat(x, 3)
print(f"x: {x}")
print(f"np.repeat(x, 3): \n{rep}")
```

## np.repeat with **axis**

```python
x = np.arange(4).reshape((2,2))
print(x,'\n')

rep = np.repeat(x, repeats=3, axis=0)
print(rep.shape, "\n", rep)
```

```python
rep = np.repeat(x, repeats=3, axis=1)
print(rep.shape, "\n", rep)
```

```python
x = np.arange(4).reshape((2,2))
print(f"x: \n{x}\n")

rep = np.repeat(x, repeats=[2, 1], axis=0) # 첫번째 원소는 2번 두번째 원소는 1번 반복
print(f"repeats=[2,1]: {rep.shape}\n{rep}")
```

```python
x = np.arange(4).reshape((2,2))
print(f"x: \n{x}\n")

rep = np.repeat(x, repeats=[2, 1], axis=1) # 첫번째 원소는 2번 두번째 원소는 1번 반복
print(f"repeats=[2,1]: {rep.shape}\n{rep}")
```

```python
import numpy as np

x = np.arange(6).reshape((2, 3))
print('x: \n',x,'\n')

rep = np.repeat(x, repeats=[2, 1, 2], axis=1)
print(rep)
```

## np.repeat with Vectors

```python
import numpy as np

row_vec = np.arange(4).reshape((1, -1))
print(row_vec, '\n' )

rep = np.repeat(row_vec, repeats=3, axis=0)
print(rep)
```

```python
col_vec = np.arange(4).reshape((-1, 1))
print(col_vec,'\n')

rep = np.repeat(col_vec, repeats=3, axis=1)
print(rep)
```

# **Overall Repetition**

## np.tile

```python
import numpy as np

a = np.arange(4)
print(a.shape,'\n',a,'\n')

tile = np.tile(a, reps=3)
print(tile.shape,'\n' ,tile)
```

## np.tile with reps

```python
import numpy as np

a = np.arange(3)
print(a.shape,'\n' ,a,'\n' )

tile = np.tile(a, reps=[1, 2])  # reps = [ 1, 2] row로 한번 col로 2번
print(tile.shape,'\n' ,tile)
```

```python
tile = np.tile(a, reps=[2, 1])
print(tile.shape,'\n' ,tile)
```

```python
tile = np.tile(a, reps=[2, 2])
print(tile.shape,'\n' ,tile)
```

```python
import numpy as np

a = np.arange(6).reshape((2,3))
print(a.shape,'\n' ,a,'\n' )

tile = np.tile(a, reps=[1,2])
print(tile.shape,'\n' ,tile)
```

```python
import numpy as np

a = np.arange(6).reshape((2,3))
print(a.shape,'\n' ,a,'\n' )

reps = np.array([3,5])
tile = np.tile(a, reps=reps)
print(f"shapes: {tile.shape} - {reps*a.shape}")
```

```python
import numpy as np

a = np.arange(4).reshape((1, -1))
tile = np.tile(a, reps=[5, 1])

print(tile)
```

```python
import numpy as np

a = np.arange(4).reshape((-1, 1))
tile = np.tile(a, reps=[1, 5])

print(tile)
```

# **Application of Repetition**

## Row-wise Case

image.png

```python
import numpy as np

x = np.arange(12).reshape((4,3))
print(f"x: {x.shape}\n{x}")
```

```python
x = x.sum(axis=0, keepdims=True) #keepdims=True => (1, 3)
print(f"x: {x.shape}\n{x}")
```

```python
x = x.repeat(repeats=3, axis=0)
print(f"x: {x.shape}\n{x}")
```

```python
x = x.T
print(f"x: {x.shape}\n{x}")
```

한줄로 표현

```python
import numpy as np

x = np.arange(12).reshape((4,3))
print(f"x: {x.shape}\n{x}\n")

y = x.sum(0, keepdims=True).repeat(3, 0).T
print(y.shape,'\n',y)
```

## Column-wise Case

image.png

```python
import numpy as np

x = np.arange(12).reshape((4,3))
print(x,'\n')

y = x.sum(axis= 1, keepdims=True).repeat(repeats=2, axis=1).T
print(y)
```

# **Making Coordinates**

## Coordinates from 1D Vectors

image.png

```python
import numpy as np

x = np.arange(-2, 3)
y = np.arange(-2, 3)

print(f"x: {x}")
print(f"y: {y}")

X = x.reshape((1,-1)).repeat(y.shape[0], axis=0)
Y = y.reshape((-1,1)).repeat(x.shape[0], axis=1)

Z = np.square(X) + np.square(Y)
print(f"Z: \n{Z}")
```

## np.mashgrid

image.png

```python
import numpy as np

x = np.arange(-2, 3)
y = np.arange(-2, 3)

X, Y = np.meshgrid(x, y)
print(X)
print(Y)
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)

X, Y = np.meshgrid(x, y)
Z = np.square(X) + np.square(Y)

fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(projection='3d')

ax.plot_wireframe(X, Y, Z)
ax.tick_params(labelsize=20)
```
