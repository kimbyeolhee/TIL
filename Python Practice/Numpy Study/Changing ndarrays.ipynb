# **Changing ndarray's shapes**

## np.reshape and ndarray.reshape

**np.reshape**

```python
import numpy as np

a = np.arange(6)
b = np.reshape(a, (2,3))

print("original ndarray: \n", a)
print("reshaped ndarray: \n", b)
```

```python
import numpy as np

a = np.arange(24)
b = np.reshape(a, (2, 3, 4))

print("original ndarray: \n", a)
print("reshaped ndarray: \n", b)
```

**ndarray.reshape**

```python
import numpy as np

a = np.arange(6)
b = a.reshape((2, 3))

print("original ndarray: \n", a)
print("reshaped ndarray: \n", b)
```

**차이점**:가독성의 입장

```python
import numpy as np

a = np.random.randint(0,100,(100,)) # 20명학생들의 5과목 점수에 관한 데이터로 가정
print(a.reshape((20,5)).mean(axis=0).max())
print(np.max(np.mean(np.reshape(a,(20,5)), axis=0)))
```

## -1 in np.reshape

```python
import numpy as np

a = np.arange(12) 
b = a.reshape((2,-1)) # row=2로 한다면 자동적으로 col=6으로 fix됨
c = a.reshape((3,-1))
d = a.reshape((-1,4))
e = a.reshape((-1,6))

print(b.shape, c.shape, d.shape, e.shape)
```

```python
a = np.arange(24)
b = a.reshape((2, 3, -1))
c = a.reshape((2, -1, 4))
d = a.reshape((-1, 3, 4))

print(b.shape, c.shape, d.shape)
```

```python
a = np.random.randint(0, 10, size=(2,2))
print(a)

row_vector = a.reshape(1, -1) # 1을 강조
col_vector = a.reshape(-1, 1)

print(row_vector.shape, col_vector.shape)
```

## np.resize and ndarray.resize

**np.resize**

shape 은 (        ) 튜플 형태로 나옴<br/>
size 는 원소의 개수: 100, 200 처럼 값으로 나옴

```python
import numpy as np

a = np.arange(6)
b = np.resize(a, (2, 3))

print("original ndarray: \n", a)
print("resized ndarray: \n", b)
```

**reshape으로는 오류가 발생**<br/>

```python
import numpy as np

a = np.arange(6)
b = np.reshape(a, (9, ))

print("original ndarray: \n", a)
print("reshaped ndarray: \n", b)
```

```python
import numpy as np

a = np.arange(10)
b = np.resize(a, (2, 3, 3))

print("original ndarray: \n", a)
print("resized ndarray: \n", b)
```

## np.flatten and np.ravel

벡터를 만드는 것

**np.flatten**

```python
import numpy as np

M = np.arange(9)
N = M.reshape((3,3))
O = N.flatten()

print(M,'\n')
print(N,'\n')
print(O,'\n')
```

3차원

```python
import numpy as np

M = np.arange(27)
N = M.reshape((3,3,3))
O = N.flatten()

print(O,'\n')

N0, N1, N2 = N[0], N[1], N[2]
print(f"N0: \n{N0}\n")
print(f"N1: \n{N1}\n")
print(f"N2: \n{N2}\n")
```

**np.ravel**

```python
import numpy as np

M = np.arange(9)
N = M.reshape((3,3))
O = N.ravel()

print(M,'\n')
print(N,'\n')
print(O,'\n')
```

# **Memory Optimization**

## copy and view of ndarrays

**copy** ; 원본에 영향을 미쳐서는 안될때<br/>
단점: 메모리 용량 차지 문제<br/>
장점: b를 수정하더라도 a에 영향을 미치지 않음<br/>
<br/>
**view** ; 원본에 영향을 미쳐도 괜찮을 때<br/>
단점: b를 수정하면 a도 수정됨 <br/> 
장점: 메모리 최적화

**view**

```python
import numpy as np

a = np.arange(5)
b = a.view()

b[0] = 100

print(a)
print(b)
```

numpy에서 **slicing**은 **view를 만드는 것!**

```python
import numpy as np

a = np.arange(5)
b = a[0:3]

b[...] = 10

print(a)
print(b)
```

**copy**

```python
import numpy as np

a = np.arange(5)
b = a.copy()

b[0] = 100

print(a)
print(b)
```

## base of ndarrays

```python
import numpy as np

a = np.arange(5)
b = a.copy()
c = a.view()
d = a[0:3]

print(b.base is a)
print(c.base is a)
print(d.base is a)
```

## APIs and copy, view

reshpae은 view를 통해 객체를 만듬<br/>
따라서 원본에도 영향을 미침

```python
import numpy as np

a = np.arange(4)
b = np.reshape(a, (2,2))

b[0,0] = 100

print(b.base is a, '\n')
print(a)
print(b)
```

resize은 copy를 통해 객체를 만듬<br/>
따라서 원본에도 영향을 미치지 않음

```python
import numpy as np

a = np.arange(4)
b = np.resize(a, (2,2))

b[0,0] = 100

print(b.base is a, '\n')
print(a)
print(b)
```

reshape는 .copy()를 통해 단점을 보완 가능

```python
import numpy as np

a = np.arange(4)
b = np.reshape(a, (2,2)).copy()

b[0,0] = 100

print(b.base is a, '\n')
print(a)
print(b)
```

ravel은 view를 통해 객체 생성

```python
import numpy as np
from numpy.random import randint

a = randint(0, 10, (2,3))
b = a.ravel()

b[0] = -10

print(b.base is a, '\n')
print(a)
print(b)
```

flatten은 copy를 통해 객체 생성

```python
import numpy as np
from numpy.random import randint

a = randint(0, 10, (2,3))
b = a.flatten()

b[0] = -10

print(b.base is a, '\n')
print(a)
print(b)
```

# **Changing ndarray's dtype**

## ndarray.astype

```python
import numpy as np

M = np.array([1, 2, 3], np.int8)
print(M.dtype)

N = M.astype(np.uint32)
O = M.astype(np.float32)
print(N.dtype)
print(O.dtype)
```

```python
import numpy as np

M = np.random.uniform(low=-10, high=10, size=(3,3))
print(M,'\n')
print(M.astype(np.int32))
```

## bool dtype ndarrays

```python
import numpy as np

bools = np.array([True, False])
print(f"bool: \n{bools}\n")

bools2ints = bools.astype(np.int)
print(f"int: \n{bools2ints}\n")

bools2floats = bools.astype(np.float)
print(f"int: \n{bools2floats}")
```

```python
import numpy as np

ints = np.array([-2, -1, 0, 1, 2])
floats = np.array([-2.5, -1.5, 0, 1.5, 2.5])

print(f"ints: \n{ints}")
print(f"floats: \n{floats}\n")

ints2bools = ints.astype(np.bool)
print(f"ints -> bools: \n{ints2bools}")

floats2bools = floats.astype(np.bool)
print(f"floats -> bools: \n{floats2bools}")
```

**BOOLEAN값**으로 바꾸어줄 때 True인 거지 **-3 값 자체**가 True인 것은 아님

```python
print(-3==True, -3==False)
print(3.14==True, 3.14==False)

print(0.==True, 0.==False)
print(1.==True, 1.==False)
```
