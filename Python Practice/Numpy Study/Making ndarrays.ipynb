# **from Python Lists**

## np.array

```python
import numpy as np

int_py = 3
float_py = 3.14

int_np = np.array(int_py)
float_np = np.array(float_py)
```

```python
print('Integer case')
print(type(int_py), type(int_np))
print(int_py, int_np, sep='-')
```

```python
print('Floating point case')
print(type(float_py), type(float_np))
print(float_py, float_np, sep='-')
```

## Making Vector ndarrays

```python
import numpy as np

vec_py = [1, 2, 3]
vec_np = np.array(vec_py)

print(type(vec_py), type(vec_np))
print(vec_py, vec_np, sep=' - ')
```

## Making Matrix ndarrays

```python
import numpy as np

mat_py = [[1, 2, 3],
          [4, 5, 6]]

mat_np = np.array(mat_py)

print(type(mat_py), type(mat_np))
print(mat_py, mat_np, sep='\n\n')
```

## Making 3rd Order Tensor ndarrays

```python
import numpy as np

tensor_py = [[[1, 2, 3],    # (2, 2, 3)
              [4, 5, 6]],
             
            [[11, 12, 13],
             [14, 15, 16]]]
tensor_np = np.array(tensor_py)
print(type(tensor_py), type(tensor_np))
print(tensor_py, tensor_np, sep='\n\n')
```

# **with Shapes**

## ndarrays with Specific Values

**np.zeros**

```python
import numpy as np

M = np.zeros(shape=(2,3))

print(M.shape)
print(M)
```

**np.ones**

```python
M = np.ones(shape=(2,3))

print(M.shape)
print(M)
```

**np.full**

```python
import numpy as np

M = np.full(shape=(2,3), fill_value=3.14)

print(M.shape)
print(M)
```

**np.empty**

```python
import numpy as np

M = np.empty(shape=(2,3))

print(M.shape)
print(M)
```

위치만 잡아 놓음

# **From Existing ndarrays**

**numpy.zeors_like()**

```python
import numpy as np

M = np.full(shape=(2,3) , fill_value=3.14)
print(M, '\n')
```

```python
zeros_like = np.zeros_like(M) # np.zeros(shape=M.shape)
ones_like = np.ones_like(M)
full_like = np.full_like(M, fill_value=100)
empty_like = np.empty_like(M)

print(f"zeros_like: \n {zeros_like}\n")
print(f"ones_like: \n {ones_like}\n")
print(f"full_like: \n {full_like}\n")
print(f"empty_like: \n {empty_like}\n")
```

# **from Random Distribution**

## from Normal Distibutions

**np.random.randn()**<br/>
평균 0, 표준편차 1

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn')

fig, ax = plt.subplots(figsize=(10,5))

random_values = np.random.randn(300)
f= ax.hist(random_values, bins=20)
```

**np.random.normal()**

```python
import numpy as np

normal1 = np.random.normal(loc=-2, scale=1, size=(200,))
normal2 = np.random.normal(loc=0, scale=2, size=(200,))
normal3 = np.random.normal(loc=3, scale=5, size=(200,))
```

```python
normal = np.random.normal(loc=[-2,0,3],
                          scale=[1,2,5],
                          size=(200,3))
```

## from Uniform Distibution

**np.random.rand()**<br/>
[0,1) 사이 값

```python
import numpy as np
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(10,5))

uniform = np.random.rand(1000)
g = ax.hist(uniform)

print(uniform.shape)
```

```python
uniform = np.random.rand(2, 3, 4)
print(uniform.shape)
```

**np.random.uniform()**<br/>
random.rand()는 0과1 사이지만<br/>random.uniform()은 원하는 low값과 high값을 설정 가능

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use("seaborn")

fig, ax = plt.subplots(figsize=(10,5))

uniform = np.random.uniform(low=-10, high=10, size=(10000,))
g = ax.hist(uniform, bins=20)

print(uniform.shape)
```
