# 2-1: Dense Layers

## 2-1-1: Shapes of Dense Layers

```python
import tensorflow as tf
from tensorflow.keras.layers import Dense

N, n_features = 1, 10
X = tf.random.normal(shape=(N, n_features))

n_neurons = 3
dense = Dense(units=n_neurons, activation='sigmoid')
Y = dense(X)

W,B = dense.get_weights()

print('==== Input/Weights/Bias====')
print('X: ', X.shape)
print('W: ', W.shape)
print('B: ', B.shape)
print('Y: ', Y.shape)
```

## Code 2-1-2: Output Calculation

```python
import tensorflow as tf
from tensorflow.keras.layers import Dense

from tensorflow.linalg import matmul
from tensorflow.math import exp

import numpy as np

N, n_feature = 4, 10
X = tf.random.normal(shape=(N, n_feature))

n_neuron = 3

# using Tensorflow 
dense = Dense(units = n_neuron, activation='sigmoid')
y_tf = dense(X)

W, B = dense.get_weights()
print("Y(Tensorflow) :\n", y_tf.numpy(),'\n')


# calculate with matrix multiplication
z = matmul(X, W) + B
y_man_matmul = 1 / (1 + exp(-z))

print("Y(with matrix multiplication :\n", y_man_matmul.numpy(),'\n')


# calculate with dot product
y_man_vec = np.zeros(shape=(N, n_neuron))

for x_idx in range(N):
  x = X[x_idx]

  for nu_idx in range(n_neuron):
    w, b = W[:, nu_idx], B[nu_idx]

    z = tf.reduce_sum(x*w) + b
    a = 1 / (1 + np.exp(-z))
    y_man_vec[x_idx, nu_idx] = a
  
print("Y(with dot product) :\n ", y_man_vec)
```
