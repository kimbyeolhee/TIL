# method

- 멤버 함수라고도 하며, 해당 클래스의 object에서만 호출 가능

```python
class Counter:
  def __init__(self):
    self.num = 0
  
  def increment(self):
    self.num += 1

  def reset(self):
    self.num = 0

  def print_current_value(self):
    print('현재값: ', self.num)

c1 = Counter()  # 객체레벨 c1 , 클래스레벨 Counter()
c1.print_current_value()
c1.increment()
c1.print_current_value()
c1.reset()
c1.print_current_value()
```

# Class Inheritance
- 의미적으로 is-a 관계를 갖는다. Student is a person

```python
class Person:
  def __init__(self, name, age):
    self.name = name
    self.age = age

  def eat(self, food):
    print(f"{self.name}은 {food}를 먹습니다")

  def sleep(self,minute):
    print(f"{self.name}은 {minute}분 동안 잡니다.")

  def work(self, minute):
    print(f"{self.name}은 {minute}동안 일합니다.")

# 상속
class Student(Person):
  def __init__(self, name, age):
    self.name = name
    self.age = age

  def work(self, minute):
    print(f"{self.name}은 {minute}동안 공부합니다.")    # method overriding

class Employee(Person):
  def __init__(self, name, age):
    self.name = name
    self.age = age

ryze = Student('Ryze', 99)
ryze.work(50)
```

# super
- 자식 클래스에서 부모 클래스의 method를 호출할 때 사용

```python
class Person:
  def __init__(self, name, age):
    self.name = name
    self.age = age

  def eat(self, food):
    print(f"{self.name}은 {food}를 먹습니다")

  def sleep(self,minute):
    print(f"{self.name}은 {minute}분 동안 잡니다.")

  def work(self, minute):
    print(f"{self.name}은 {minute}동안 준비합니다.")

# 상속
class Student(Person):
  def __init__(self, name, age):
    self.name = name
    self.age = age

  def work(self, minute):
    super().work(minute)     # super
    print(f"{self.name}은 {minute}동안 공부합니다.")    # method overriding

class Employee(Person):
  def __init__(self, name, age):
    self.name = name
    self.age = age

ryze = Student('Ryze', 99)
ryze.work(50)
```

# 클래스 연습

```python
class Point:
  def __init__(self, x, y):
    self.x = x
    self.y = y

  def __str__(self):
    return f'{self.x}, {self.y}'

  def __add__(self, pt):
    new_x = self.x + pt.x
    new_y = self.y + pt.y
    return Point(new_x, new_y)

  def __sub__(self, pt):
    new_x = self.x - pt.x
    new_y = self.y - pt.y
    return Point(new_x, new_y)


p1 = Point(3, 4)
p2 = Point(2, 7)

p3 = p1 + p2
print(p3)

p4 = p1 - p2
print(p4)
```
