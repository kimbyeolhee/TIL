# generate Tensor

```python
import torch
```

```python
# create a torch.Tensor object with data
V_data = [1., 2., 3.]
V = torch.Tensor(V_data)
print(V)

M_data = [[1, 2, 3], [4, 5, 6]]
M = torch.Tensor(M_data)
print(M)

T_data = [[[1,2], [3,4]],
          [[5,6], [7,8]]]
T = torch.Tensor(T_data)
print(T)
```

```python
# torch.ones
a = torch.ones(3)
print(a)

b = torch.ones(2, 3)
print(b)
```

```python
# torch.zeros
a = torch.zeros(4)
print(a)

b = torch.zeros(3, 2, 4)
print(b)
```

```python
# 초기화하면서 현재 메모리에 존재하는 값
x = torch.empty(5,3)
print(x)

# 랜덤 숫자로 채움
x = torch.rand(5, 3)
print(x)

x = torch.rand(4, dtype=torch.double)
print(x)

x = torch.rand(2, 3, dtype=torch.float)
print(x)
```

```python
# size
print(x.size())
```

# Indexing Tensor

```python
temp = torch.Tensor([0, 1, 2, 3, 4, 5 ,6 ,7])

print(temp[1:4])
print(temp[1:7:2])
```

```python
temp = torch.Tensor([[1,2,3],[4,5,6]])

print(temp)
print(temp[1:])
print(temp[1:, 1:])
print(temp[:, 0])
```

# Tensor element types

```python
float_points = torch.ones(5,2, dtype=torch.float32)
int_points = torch.zeros(3, dtype=torch.int64)

print(float_points)
print(int_points)
```

```python
default_points = torch.Tensor([2,2])
print(default_points)
```

```python
# find out type
print(float_points.dtype)
print(int_points.dtype)
print(default_points.dtype)
```

```python
# change dtype in specific
to_int = float_points.int()
print(to_int.dtype)

# using 'to'
to_int_2 = float_points.to(torch.int64)
print(to_int_2.dtype)

to_float = int_points.to(torch.float16)
print(to_float.dtype)
```

# Tensor Shape and Reshape

```python
a = torch.Tensor([[1,2,3],[4,5,6]])

print(a)
print(a.shape)
```

```python
# size
print(a.size())
```

```python
# reshape
a_reshape = a.reshape(3,2)
print(a_reshape)

a_reshape = a.reshape(-1,2)
print(a_reshape)
```

```python
# if shape is not compatible with the current shape -> ERROR
a_reshape = a.reshape(4,2)
print(a_reshape)
```

```python
# view : 원본 input과 동일한 저장이 필요할 경우
x = torch.randn(4,4)
print(x)
print(x.size(),'\n')

y = x.view(16)
print(y)
print(y.size(),'\n')

z = x.view(-1,8)
print(z)
print(z.size(),'\n')
```

# Transpose

```python
x = torch.randn(3)
print(x)
print(torch.t(x),'\n')

y = torch.randn(2,3)
print(y)
print(torch.t(y))
```

```python
a = torch.randn(2, 3)
print(a)
print(torch.transpose(a, 0, 1))
```

```python
# complex case
ex = torch.randn(64, 128, 728)
print(ex.shape)
transposed = torch.transpose(ex, 1, 2)
print(transposed.shape)
```

# Operation

```python
x = torch.ones(1, 2)
y = torch.ones(5, 2)
print(y)
#
print(x+y)
#
print(torch.add(x,y))
#
result = torch.empty(5, 2)
torch.add(x, y, out=result)
print(result)
#
print(y.add_(x))
```
